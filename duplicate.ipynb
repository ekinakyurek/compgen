```python
import json
import os
```

```python
originaldata = "data/SIGDataSet.large/"
copydata = "data/SIGDataSet.large.copy2/"
langs = ("turkish","spanish","swahili")
```

```python
def duplicate(original_root, copy_root, langs):
    for lang in langs:
        datapath = original_root + lang + "/"
        copypath = copy_root + lang + "/"
        os.makedirs(copypath,exist_ok=True)
        vocab = json.load(open(datapath+"vocab.json",'r'))
        json.dump(vocab, open(copypath + "vocab.json",'w'))
        PST, FUT = vocab["PST"], vocab["FUT"]
        
        for hints in (4,8,16):
            for seed in (0,1,2,3,4):
                neighborhoods = json.load(open(datapath + f"neighborhoods.hints-{hints}.{seed}.json", 'r'))
                seqs = json.load(open(datapath +  f"seqs.hints-{hints}.{seed}.json", 'r'))
                splits = json.load(open(datapath +  f"splits.hints-{hints}.{seed}.json", 'r'))
                trn_ids = splits['train']
                rare_ids = [ids for ids in trn_ids if (PST in seqs[ids]) or (FUT in seqs[ids])]
                print(f"length of seq: {len(seqs)}")
                print(f"length of rares: {len(rare_ids)}")
                assert len(rare_ids)==2*hints
                for t in range(22):
                    for rare_id in rare_ids:
                        new_id = len(seqs)
                        seqs.append(seqs[rare_id])
                        trn_ids.append(new_id)
                        print(f"new id {new_id}, rare id {rare_id}")
                        neighborhoods[str(new_id)] = neighborhoods[str(rare_id)]
                json.dump(neighborhoods, open(copypath + f"neighborhoods.hints-{hints}.{seed}.json",'w'))
                json.dump(seqs, open(copypath +  f"seqs.hints-{hints}.{seed}.json",'w'))
                json.dump(splits, open(copypath +  f"splits.hints-{hints}.{seed}.json",'w'))
                
```

```python
duplicate(originaldata, copydata, langs)
```
